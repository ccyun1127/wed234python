```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=28387&md5_url=a3486ce502e239621e696782dfb49825')
```

```python
x1 = df[(df.項目=='貸款人次')].公立大專校院
x2 = df[(df.項目=='貸款人數')].公立大專校院
x3 = df[(df.項目=='貸款金額（元）')].公立大專校院
plt.plot(x2, label='Student Amount')
plt.plot(x1, label='Total Times')
plt.plot(x3, label='Debt Money Total')
plt.legend()
plt.show()
```

```python
x1 = df[(df.項目=='貸款金額（元）')].公立大專校院
x2 = df[(df.項目=='貸款金額（元）')].私立大專校院
x3 = df[(df.項目=='貸款金額（元）')].公立高中職
x4 = df[(df.項目=='貸款金額（元）')].私立高中職

plt.plot(x3, label='Public High school')
plt.plot(x4, label='Private High School')
plt.plot(x1, label='Public College')
plt.plot(x2, label='Private College')



plt.legend()
plt.show()
```
